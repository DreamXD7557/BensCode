```python
import random

# Function to generate a multiplication question
def ask_question():
    num1 = random.randint(1, 12)  # Random number between 1 and 12
    num2 = random.randint(1, 12)  # Random number between 1 and 12
    correct_answer = num1 * num2  # Correct answer
    user_answer = int(input(f"What is {num1} x {num2}? "))  # Get user's answer
    return user_answer == correct_answer  # Return True if the answer is correct

def quiz():
    score = 0 # Initialize score

    for i in range(5):
        print(f"Question {i + 1}:")
        if ask_question():
            print("Correct!\n")
            score += 1
        else:
            print("Incorrect.\n")

    print(f"You got {score} out of 5 correct!")

# Run the quiz
quiz()
```

```python
import random

# Function to generate a multiplication question based on level
def ask_question(level):
    if level == 'easy':
        # Easy level with 2, 5, and 10 times tables
        times_table = random.choice([2, 5, 10])  # Randomly choose 2, 5, or 10
        num = random.randint(1, 10)  # Random number between 1 and 10
        correct_answer = times_table * num
        user_answer = int(input(f"Easy level - What is {times_table} x {num}? "))

    elif level == 'medium':
        # Medium level with numbers from 11 to 20
        num1, num2 = random.randint(11, 20), random.randint(11, 20)
        correct_answer = num1 * num2
        user_answer = int(input(f"Medium level - What is {num1} x {num2}? "))

    elif level == 'hard':
        # Hard level with numbers from 21 to 30
        num1, num2 = random.randint(21, 30), random.randint(21, 30)
        correct_answer = num1 * num2
        user_answer = int(input(f"Hard level - What is {num1} x {num2}? "))

    return user_answer == correct_answer  # Return True if the answer is correct

def quiz():
    print("Welcome to the Multiplication Quiz!")
    # User picks the difficulty level
    level = input("Choose your level (easy, medium, hard): ").lower()

    if level not in ['easy', 'medium', 'hard']:
        print("Invalid level chosen. Please restart and choose between 'easy', 'medium', or 'hard'.")
        return

    score = 0  # Initialize score

    # Ask 5 questions based on the chosen level
    for i in range(1, 6):
        print(f"\nQuestion {i}:")
        if ask_question(level):
            print("Correct!\n")
            score += 1
        else:
            print("Incorrect.\n")

    print(f"\nYou got {score} out of 5 correct!")

# Run the quiz
quiz()
```

```python
import random

# Function to generate a multiplication question based on level
def ask_question(level):
    # Set the range of numbers based on the difficulty level
    if level == 1:
        num1, num2 = random.randint(1, 12), random.randint(1, 12)  # Level 1: 1-12
    elif level == 2:
        num1, num2 = random.randint(13, 20), random.randint(13, 20)  # Level 2: 13-20
    elif level == 3:
        num1, num2 = random.randint(21, 30), random.randint(21, 30)  # Level 3: 21-30
    elif level == 4:
        num1, num2 = random.randint(31, 40), random.randint(31, 40)  # Level 4: 31-40
    else:
        num1, num2 = random.randint(41, 50), random.randint(41, 50)  # Level 5: 41-50

    correct_answer = num1 * num2
    user_answer = int(input(f"Level {level} - What is {num1} x {num2}? "))
    return user_answer == correct_answer  # Return True if the answer is correct

def quiz():
    score = 0  # Initialize score

    # Loop through 5 levels of difficulty
    for level in range(1, 6):
        print(f"\nStarting Level {level}...")
        if ask_question(level):
            print("Correct!\n")
            score += 1
        else:
            print("Incorrect.\n")

    print(f"\nYou got {score} out of 5 correct!")

# Run the quiz
quiz()
```
